```python
import requests
url = "https://www.kebhana.com/cms/rate/wpfxd651_01i_01.do"
```

```python
payload = {"ajax" : "true",
            "curCd" : "",
            "pbldDvCd" : "0",
            "pbldSqn" : "",
            "hid_key_data" : "",
            "inqStrDt" : "20230307",
            "inqKindCd" : "1",
            "hid_enc_data" : "",
            "requestTarget" : "searchContentDiv",}
```

```python
r = requests.post(url, data=payload)
```

```python
hana = pd.read_html(r.text)[0]
```

```python
def get_exchange(date_, currency, type_=1):
    """
    date_ : 날짜, 날짜는 yyyy-mm-dd 형식으로 입력 
    currency: 화폐 코드 
    type_ : 0 사실때 , 1 파실때 
    """
    payload['inqStrDt'] = date_.replace("-","")
    r  = requests.post(url, data=payload)
    hana = pd.read_html(r.text)[0]
    hana.columns = ["_".join(list(set(x))) for x in hana.columns]
    if type_ == 0:
        return hana.loc[hana.통화.str.find(f"{currency}".upper()) > -1,  "사실 때_현찰_환율"].values[0]
    elif type_ == 1:
        return hana.loc[hana.통화.str.find(f"{currency}".upper()) > -1,  "현찰_파실 때_환율"].values[0]
    else:
        return -1
```
