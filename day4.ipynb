* KBO 선수 BMI 지수 구하고 최고 5인 추출

```python
import re
```

```python
p = re.compile("[0.9]+") # 숫자를 알려줌, 정규식 정규표현식
```

```python
p.findall('187cm/94kg')
```

```python
height, weight = p.findall('187cm/94kg')
```

```python
for x in kbo :  # [:-2] 이 자리만큼 빼줌
    height, weight = x['신장/체중'].split("/")
    bmi = int(weight[:-2]) / (int(height[:-2]) /100) ** 2
    x['bmi'] = round(bmi, 2) # round는 소숫점 자리 끊어주는
    x['나이'] = 2023 - int(x['생년월일'][:4]) #date.today().year
```

```python
sorted({"-".join([x['선수명'],x['등번호']]) : x['bmi'] for x in kbo}.items(), key=lambda x: x[1], reverse=True)
```

```python
salary = {}
for play in kbo:
    if play['연봉'][-2:] == '만원':
        if play['Team'] in salary:
            salary[play['Team']] += [int(play['연봉'][:-2])]
        else:
            salary[play['Team']] = [int(play['연봉'][:-2])]
```

```python
sorted({key : ( sum(value)/len(value)) for key,value in salary.items() }.items(),
       key=lambda x : x[1], reverse=True) # 평균값, 내림차 순
```

```python
import numpy as np
sorted({key : (np.median(value)) for key,value in salary.items() }.items(),
       key=lambda x : x[1], reverse=True)
```

```python
url = "http://data.ex.co.kr/openapi/locationinfo/locationinfoRest?key=8097256368&type=xml&numOfRows=1000&pageNo=1"
```

```python
r = requests.get(url)
```

```python
import xml.etree.ElementTree as ET
root = ET.fromstring(r.text) #파시브 할 수 있게 넘겨줘 xml
```

```python
total = []
items = root.iter(tag = 'list') #리스트 반복되니까 이걸 아이템에 담아
for element in items :
    tmp = {}
    for x in element :
        tmp[x.tag] = x.text #태그와 텍스트를 키와 밸류값으로 total에 담아
    total.append(tmp)
```
