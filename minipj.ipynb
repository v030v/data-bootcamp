```python
# rename - 이름 변경, reindex - 인덱스 순서 수정
dust = dust.rename(index={'총계':'계','서울특별시':'서울','부산광역시':'부산',
                            '인천광역시':'인천','대구광역시':'대구','광주광역시':'광주',
                            '대전광역시':'대전','울산광역시':'울산','경기도':'경기','강원도':'강원',
                            '충청북도':'충북','충청남도':'충남','전라북도':'전북','전라남도':'전남',
                            '경상북도':'경북','경상남도':'경남','제주특별자치도':'제주','세종특별자치시':'세종'})
dust = dust.reindex(['계','서울','부산','인천','대구','광주',
              '대전','울산','경기','강원','충북','충남','전북',
              '전남','경북','경남','제주','세종'])
```

```python
age = pd.read_pickle("./mini-1/age.pkl")
```

```python
age.insert(0,'',range(1,22))
```

```python
age = age.set_index('')
```

```python
age2 = age.set_index('나이').groupby(['나이']).sum()
```

```python
# 남녀 2-30대의 값을 더해서 묶어
young = pd.concat([age.iloc[4:6,:],age.iloc[14:16,:]]).groupby(['나이']).sum()
```

```python
# young의 컬럼 값을 다 더해
young.sum()
```

```python
# '청년 총계'라는 인덱스를 추가하고 young.sum() 값을 넣어줘
young_tot = young.loc['청년 총합'] = young.sum()
```

```python
young
```

```python
old = pd.concat([age.iloc[8:10,:],age.iloc[18:20,:]]).groupby(['나이']).sum()
```

```python
old.sum()
```

```python
old_tot = old.loc['노인 총합'] = old.sum()
```

```python
old
```

```python
tot = pd.concat([young,old],axis=0)
```

```python
tot
```

```python
fig = plt.figure(figsize=(25,6)) # 캔버스 사이즈 설정
ax = fig.add_subplot() # 그림 뼈대(프레임) 생성
```

```python
plt.plot(tot.columns, young_tot, marker='o', label='청년 총합',color='red')
plt.plot(tot.columns, old_tot, marker='o', label='노인 총합',color='blue')

plt.title('연령대별 환자 수', fontsize = 15) # 타이틀 설정
plt.xticks(tot.columns, rotation=45) # x값 글씨 기울기
plt.locator_params(axis='x',nbins =len(tot.columns)/2) # x값 수치 갯수?
plt.show()
```
